```python
import requests
import pandas as pd

def history(symbol, interval = '1d', limit = '500'):
    """
    Get Cryptocurrency market data from Binance API
    interval take: 1s, 1m, 1h, 1d, 1w
    """
        # Send request to Binance api
        url = f'https://api.binance.com/api/v3/klines?symbol={symbol}&interval={interval}&limit={limit}'
        response = requests.get(url)
        data = response.json()
        
        # Clean dataset
        df = pd.DataFrame(data)
        df.columns = ['open_time','Open', 'High', 'Low', 'Close', 'Volume','close_time', 'QuoteVolume', 'Num_trades','taker_base_vol', 'taker_quote_vol', 'ignore']
        df = df.drop(['open_time','close_time','taker_base_vol','taker_quote_vol','ignore'],axis=1)
        df = df.astype(float)
        return df
```

```python
history('BTCUSDT','1h')
```

```python
history('BTCUSDT','1w')
```
